### Le calcule de la durée pour chaque trajet en minutes

```python
import numpy as np

dprt = tickets["departure_ts"]
arv = tickets["arrival_ts"]

t1 = pd.to_datetime(dprt)
t2 = pd.to_datetime(arv)

deff = t2-t1
traject_duration_min = deff /np.timedelta64(1,'s')/60

tickets["duration"] = traject_duration_min

min_duration = traject_duration_min.min()
print ("Duree minimale d'un projet :",min_duration)

max_duration = traject_duration_min.max()
print ("Duree maximale d'un trajet :",max_duration)

mean_duration = traject_duration_min.mean()
print ("La moyenne des trajert",mean_duration)
```

## L'ajout des coordonnées des stations de départs et d'arrivées au dataFrame ticket_data

```python
colums_needed = stations.loc[:,["id","latitude","longitude"]] # récuperation des columns dont j'ai besion pour les coordonnées des stations de départ

merg_o = pd.merge(tickets,colums_needed,left_on = "o_station",right_on = "id",indicator = True) # faire le merge (dèpart)
rename_col_o = merg_o.rename(columns={"latitude":"o_latitude", "longitude":"o_longitude"}) #renommer les columns

merg_d = pd.merge(rename_col_o,colums_needed,left_on = "d_station",right_on = "id") # faire le merge (arriver)
rename_col_d = merg_d.rename(columns={"latitude":"d_latitude", "longitude":"d_longitude"})# renommer les columns
```

## Calcule des distances  en utilisant la library geopy

```python
from geopy.distance import vincenty

def distance_calc (row):
    start = (row['o_latitude'], row['d_longitude'])
    stop = (row['d_latitude'], row['d_longitude'])

    return vincenty(start, stop).km # La fonction pour calculer les distances en KM

rename_col_d['distance'] = rename_col_d.apply (lambda row: distance_calc (row),axis=1) #appliquer la fonction sur le dataframe
```

## Création des labels pour les distances

```python
def which_label(row):
    distance = row['distance']
    
    if distance <= 200:
        return "0-200"
    elif distance >= 201 and distance <= 800:
        return "201-800"
    elif distance >= 801 and distance <= 2000:
        return "801-2000"
    elif pd.isnull(distance):
        return "Unknown"
    else:
        "+2000"
```

```python
labels = rename_col_d.apply(which_label, axis = 1)

rename_col_d["dis_label"] = labels
rename_col_d.head() 
```

```python
rename_col_d.loc[300:350,["distance","dis_label"]] #Vérification des labels
```

## La moyenne des prix et durées selon les labels des distances

```python
stat = rename_col_d.pivot_table(index = 'dis_label', values = ["duration","price_in_cents"], aggfunc = np.mean)

print stat
```
